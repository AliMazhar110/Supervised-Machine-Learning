# Regression

### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns",None)
pd.set_option("display.precision",2)
```

### Importing data

```python
data1 = pd.read_csv("ex1data1.csv")
data1.head()
```

```python
data_array = data1.iloc[:,:].values
print(data_array)
```

```python
x = []
y = []
for lst in data_array:
    s = lst[0]
    sep_lst = s.split(",")
    x.append(float(sep_lst[0]))
    y.append(float(sep_lst[1]))
```

```python
x,y
```

```python
#converting list to array
x = np.array(x)
y = np.array(y)
x,y
```

```python
#converting 1d array to 2d array
x = x.reshape(-1,1)
y = y.reshape(-1,1)
x,y
```

### Visualization of data

```python
plt.figure(figsize=(20,15))
plt.scatter(x,y,color="red",marker="x",s=100)
plt.xlabel("Population in 10,000s")
plt.ylabel("Profit in 10,000s")
plt.title("Population vs Profit")
plt.show()
```

## Using sklearn

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x_train, y_train, color="red",s=50)
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.xlabel("Population in 10,000s")
plt.ylabel("Profit in 10,000s")
plt.title("Population vs Profit")
plt.show()
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x_test, y_test, color="red",s=50)
plt.plot(x_test, y_pred, color="blue")
plt.xlabel("Population in 10,000s")
plt.ylabel("Profit in 10,000s")
plt.title("Population vs Profit")
plt.show()
```

```python
r_score = r2_score(y_test,y_pred)
print("r2 score =",r_score)
```

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
print("MSE =",mse,",MAE =",mae)
```

## Without Using sklearn

```python
x_ws = np.c_[np.ones(len(x_train)), x_train]
y_ws = y_train
theta = np.zeros((2,1))
iterations = 1500
alpha = 0.01
```

```python
x_ws,theta
```

```python
x_ws.shape, theta.shape,y_ws.shape
```

```python
def hypothesis(X, theta, y):
    return (X.dot(theta) - y)
```

```python
def cost_function(X,y,theta):
    m = len(y)
    J = (hypothesis(X,theta,y)**2)*(1/(2*m))
    return np.sum(J)
```

```python
print(cost_function(x_ws,y_ws,theta))
```

```python
def gradient_descent(X,y,theta,alpha,iterations):
    m = len(y)
    J_history = []
    J_history.append(cost_function(X,y,theta))
    for i in range(iterations):
        temp = X.T.dot(((hypothesis(X,theta,y))))
        print(temp)
        theta = theta - alpha*(1/m)*temp
        J_history.append(cost_function(X,y,theta))
    return theta,J_history
```

```python
theta,J_history = gradient_descent(x_ws,y_ws,theta,alpha,iterations)
print("theta =",theta)
```

```python
print(J_history[-1])
```

```python
print(f"Final Equation => y = {theta[0][0]:.2f} + {theta[1][0]:.2f} * x")
```

```python
plt.figure(figsize=(20,10))
plt.plot(J_history,color="blue")
plt.xlabel("Iterations")
plt.ylabel("J(ↀ)")
plt.title("Cost Function vs Iterations")
plt.show()
```

```python
plt.figure(figsize=(20,15))
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,x_ws.dot(theta),color="blue")
plt.xlabel("Population in 10,000s")
plt.ylabel("Profit in 10,000s")
plt.show()
```
